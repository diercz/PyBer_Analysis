```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot bar chart
plt.bar(x_axis, y_axis, label='Boston', color= "green")

#create labels
plt.xlabel("Date")
plt.ylabel("Fare($)")

#create a title
plt.title("PyBer Fare by Month")

#create legend
plt.legend()
```

```python
#creating inverted bar chart
plt.barh(x_axis, y_axis, color="magenta", label ='Woodbury')

#invert Y axis to reorder with January first
plt.gca().invert_yaxis()

#add title and axis labels
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")

plt.legend()

```

```python
#create vertical bar chart with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black", label='New York')

plt.legend()
```

```python
#create horizontal bar chart with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')

plt.legend()

#invert the y axis to reorder month
plt.gca().invert_yaxis()

#labels
plt.title= "PyBer Fare by Month"
plt.xlabel="Fare($)"
plt.ylabel="Date"
```

```python
#scatter plots
plt.scatter(y_axis, x_axis, color="red", label='Chicago')


#invert y axis to relabel months
plt.gca().invert_yaxis()

plt.xlim(0,50)

#label
plt.title="Pyber Fare by Month"
plt.xlabel="Month"
plt.ylabel="Fare($)"

plt.legend()
```

```python
# Create a Scatter Plot Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a Bubble Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis, s=[i*5 for i in y_axis], color='skyblue', alpha=0.8, edgecolors="black", linewidth=2, label="Boston")
plt.legend()

plt.xlim(0,50)

plt.gca().invert_yaxis()

#create labels
plt.xlabel("Fare($)")
plt.ylabel("Date")

#create a title
plt.title("PyBer Fare by Month")
plt.show()
```

```python
#assign colors
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

#Explode a piece of the pie
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)

#resize to fit all values with pie slice
plt.subplots(figsize=(8,8))


plt.pie(y_axis, 
        labels=x_axis, 
        explode=explode_values, 
        colors=colors,
        autopct='%.1f%%')

plt.gca().invert_yaxis()

plt.show()


```

```python
# Create a Pie Chart Using the Object-Oriented Interface

#assign colors
colors = ["slateblue", "seagreen", "lavender", "forestgreen", "chartreuse", "darkseagreen", "firebrick", "peru", "gold", "indianred", "crimson", "dodgerblue"]

#Explode a piece of the pie
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)

fig, ax = plt.subplots(figsize=(6,6))
ax.pie(y_axis,
       labels=x_axis,
      explode=explode_values, 
        colors=colors,
        autopct='%.1f%%',
      shadow=True)

plt.show()
```

```python
# create the plot
plt.plot(x_axis, y_axis, label="Boston", marker="d", color= "green", linewidth=2)

# Create labels for the x and y labels
plt.xlabel("Date")
plt.ylabel("Fare($)")

# set the y limits
plt.ylim(0, 45)

# create title
plt.title("PyBer Fare by Month")

#add legend
plt.legend()

# add a grid
plt.grid()

# how to use xlim()
plt.xlim(0,12)

# how would you change the fig size?
plt.figure(figsize=(20,10))
```
